```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
from PIL import Image, ImageOps
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense
from keras import optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
```

```python
# Load data
images = []
ages = []

for i in os.listdir('utkface_aligned_cropped/crop_part1/')[0:8000]:
    split = i.split('_')
    ages.append(int(split[0]))
    images.append(Image.open('utkface_aligned_cropped/crop_part1/' + i))

images = pd.Series(list(images), name='Images')
ages = pd.Series(list(ages), name='Ages')

df = pd.concat([images, ages], axis=1)
```

```python
# Display initial data samples
display(df['Images'][0])
print(df['Ages'][0])
```

```python
# Distribution plot of ages
sns.set_theme()
sns.histplot(df['Ages'], kde=True, bins=30)
```

```python
# Under-sample ages below 4
under4s = df[df['Ages'] <= 4].sample(frac=0.3)
df = pd.concat([df[df['Ages'] > 4], under4s], ignore_index=True)
```

```python
# Limit ages to below 80 for the dataset
df = df[df['Ages'] < 80]
sns.histplot(df['Ages'], kde=True, bins=20)
```

```python
df.reset_index(drop=True, inplace=True)

# Image processing and resizing
x = []
y = df['Ages'].values  # Only ages as labels

for i in range(len(df)):
    df.loc[i, 'Images'] = df.loc[i, 'Images'].resize((200, 200), Image.LANCZOS)  # Updated with loc to avoid chained assignment
    ar = np.asarray(df.loc[i, 'Images'])
    x.append(ar)

x = np.array(x)
```

```python
# Train-test split for age data
x_train_age, x_test_age, y_train_age, y_test_age = train_test_split(x, y, test_size=0.2, stratify=y)
```

```python
# Define age prediction model
agemodel = Sequential()
agemodel.add(Conv2D(32, (3,3), activation='relu', input_shape=(200, 200, 3)))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Conv2D(64, (3,3), activation='relu'))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Conv2D(128, (3,3), activation='relu'))
agemodel.add(MaxPooling2D((2,2)))
agemodel.add(Flatten())
agemodel.add(Dense(64, activation='relu'))
agemodel.add(Dropout(0.5))
agemodel.add(Dense(1, activation='relu'))  # Single output for age regression


agemodel.compile(loss='mean_squared_error', optimizer=optimizers.Adam(learning_rate=0.0001))
```

```python
# Data augmentation for age model
datagen = ImageDataGenerator(
    rescale=1./255., width_shift_range=0.1, height_shift_range=0.1, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Train and test data generators for age model
train1 = datagen.flow(x_train_age, y_train_age, batch_size=32)
test1 = test_datagen.flow(x_test_age, y_test_age, batch_size=32)
```

```python
# Train age model
history1 = agemodel.fit(train1, epochs=50, shuffle=True, validation_data=test1)
```
